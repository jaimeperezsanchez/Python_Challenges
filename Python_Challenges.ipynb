```python
from math import sqrt

def resolution(pul, pix_ancho, pix_alto):
    hip = sqrt(pix_ancho**2 + pix_alto**2)
    anch = int((pul/hip) * pix_ancho * 25.4)
    alto = int((pul/hip) * pix_alto * 25.4)
    return (anch, alto)

N = int(input())
output = []
for _ in range(N):
    inp = input().split()
    inp = list(map(float, inp))
    a, b = resolution(inp[0], inp[1], inp[2])
    output.extend([a,b])
#print
for i in range(0,len(output),2):
  print(output[i],output[i+1])
```

```python
Symbols = "ABCDEFGHIJKLMNOPQRSTUVWXYZ "

def cesar_cipher(message, key):
    cipher = ""
    for word in message:
        i = (Symbols.index(word) + key) % len(Symbols)
        cipher += Symbols[i]
    return cipher

def probable_words(message):
    words = ["CESAR", "LEGIONES", "GALIA", "VERCINGETORIX", "GERGOVIA", "FRENTE"]
    for word in words:
        if word in message:
            return True
    return False

n = int(input())
for N in range(n):
    message = input()
    for key in range(len(Symbols)):
        cipher = cesar_cipher(message, key)
        if probable_words(cipher):
            print(cipher)
            break
```

```python
from math import sqrt

def divs(number):
    factors = set()
    for p in range(1, (int(sqrt(number))+1)):
      if number % p == 0:
        q = number//p
        candidate = tuple([p,q])
        factors.add(candidate)
    best = number + 2
    for i in factors:
      sum_i = sum(i)
      if sum_i < best:
        best = sum_i
        best_p = i[0]
        best_q = i[1]
    return min(best_p,best_q), max(best_p,best_q)

N = int(input())
output = []

for _ in range(N):
    number = int(input())
    p, q = divs(number)
    output.extend((p,q))
for i in range(0,len(output),2):
  print(output[i],output[i+1])
```

```python
from math import sqrt

# https://codereview.stackexchange.com/questions/255026/determine-whether-a-number-and-its-square-are-both-the-sum-of-two-squares
# https://www.geeksforgeeks.org/check-whether-number-can-represented-sum-two-squares/

def is_sum_of_two_squares(n):
    """Check whether n can be written as i*i + j*j."""
    # Assume 1 <= i <= j
    # Then n = i*i + j*j becomes
    #      n <= 2*j*j which implies j >= sqrt(n/2)
    # And  n >= 1*1 + j*j which implies j <= sqrt(n-1)
    for j in range(int(sqrt(n/2)), 1 + int(sqrt(n-1))):
        i2 = n - j * j
        i = int(sqrt(i2))
        if i * i == i2 and i > 0:
            return True
    return False

def SieveOfEratosthenes(n):
    prime = [True]*(n+1)
    prime[0]= False
    prime[1]= False
    for p in range(2,int(sqrt(n))+1):
        # If prime[p] is not changed, then it is a prime
        if (prime[p] == 1):
            # Update all multiples of p
            for i in range(p ** 2, n + 1, p):
                prime[i] = False
    prime_list = [i for i, n in enumerate(prime) if n == 1]
    return prime_list

N = int(input())
output = []
number_max = 10000
prime_list = SieveOfEratosthenes(number_max)
for _ in range(N):
    number = int(input())
    if number == 1:
        output.append(0)
    elif number == 2:
        output.append(1)
    else:
        if not(number in prime_list):
            output.append(0)
        else:
            # is especial
            if is_sum_of_two_squares(number):
                output.append(2)
            # is prime
            else: 
                output.append(1)

# NO: 0
# PRIMO: 1
# ESPECIAL: 2
for i in range(len(output)):
    if output[i] == 0:
        print('NO')
    elif output[i] == 1:
        print('PRIMO')
    else:
        print('ESPECIAL')
```
